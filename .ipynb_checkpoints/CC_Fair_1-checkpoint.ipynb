## **<font color="OrangeRed"> Ejemplo de una clase </font>**
En programación orientada a objetos, una clase es una plantilla o modelo que define las características y comportamientos de un conjunto de objetos. Un objeto es una instancia específica de una clase, que tiene los atributos y métodos definidos en esa plantilla.

Imagina que una clase es como el diseño arquitectónico de una casa: describe cómo debería ser una casa, pero no es una casa en sí. Cuando se construye una casa basada en ese diseño, entonces tenemos un objeto que sigue el plano original.

En el contexto de una clase universitaria, podemos pensar en una clase de programación que modele el comportamiento y la información de una asignatura. Esta clase podría incluir atributos como los nombres de los estudiantes, sus equipos de trabajo y sus calificaciones. Además, podría tener métodos para agregar estudiantes, asignar calificaciones y mostrar la información de los participantes.

El siguiente paso es iniciar la clase, lo que en términos de programación significa crear una instancia de esa clase y guardarla en una variable, para poder manipularla y utilizar sus métodos.

```python
# Definición de la clase ClaseUniversitaria
class ClaseUniversitaria:
    def __init__(self, nombre_clase):
        self.nombre_clase = nombre_clase
        self.estudiantes = {}  # Diccionario para almacenar nombres de estudiantes y sus equipos
        self.calificaciones = {}  # Diccionario para almacenar nombres de estudiantes y sus calificaciones
        print(f"Clase '{self.nombre_clase}' iniciada.")

    def agregar_estudiante(self, nombre_estudiante, equipo):
        # Agrega un estudiante con su equipo de trabajo
        self.estudiantes[nombre_estudiante] = equipo
        self.calificaciones[nombre_estudiante] = None  # Inicialmente, no hay calificación
        print(f"Estudiante '{nombre_estudiante}' agregado al equipo '{equipo}'.")

    def asignar_calificacion(self, nombre_estudiante, calificacion):
        # Asigna una calificación a un estudiante
        if nombre_estudiante in self.calificaciones:
            self.calificaciones[nombre_estudiante] = calificacion
            print(f"Calificación {calificacion} asignada a '{nombre_estudiante}'.")
        else:
            print(f"Estudiante '{nombre_estudiante}' no encontrado en la clase.")

    def mostrar_informacion(self):
        # Muestra la lista de estudiantes, equipos y calificaciones
        print(f"Información de la clase '{self.nombre_clase}':")
        for estudiante, equipo in self.estudiantes.items():
            calificacion = self.calificaciones[estudiante]
            print(f"Estudiante: {estudiante}, Equipo: {equipo}, Calificación: {calificacion}")

# Paso 1: Iniciar la clase
clase = ClaseUniversitaria("Física del Clima")

# Agregar algunos estudiantes y asignar equipos
clase.agregar_estudiante("Ana Pérez", "Equipo 1")
clase.agregar_estudiante("Carlos López", "Equipo 2")

# Asignar calificaciones a los estudiantes
clase.asignar_calificacion("Ana Pérez", 95)
clase.asignar_calificacion("Carlos López", 88)

# Mostrar la información de la clase
clase.mostrar_informacion()
```

Es importante mencionar en este punto que FaIR es un modelo de cajas, en este caso una "layer" se refiere a las capas, la primera es la superficie, la segunda es la primer capa océanica y la tercera es el océano profundo. Esto en su definición por default son 3 "layers". Los "gasboxes" son las cajas atmosféricas donde se guardan los gases y donde se resuelve la temperatura. La caja más inferior es la superficial. 



```python
print(f.layers)
print(f.gasboxes)
```

## **<font color="ForestGreen"> Paso 2 Tiempo de simulacion </font>**

En FaIR el tiempo es la única dimensión y por lo tanto, técnicamente podríamos hacer la simulación que queramos en el tiempo que queramos. De manera general, vamos a querer hacer simulaciones que sean parecidas a las del IPCC. 

Hay dos indicadores diferentes del tiempo en este modelo: `timebound` y `timepoint`. 
El intervalo de tiempo `timebound` son las coordenadas del tiempo que rodean o que se encuentran alrededor de un paso de tiempo, son fotos instantáneas de lo que está sucediendo. El punto en el tiempo, `timepoint` es lo que sucede en el intervalo de tiempo. Es decir, cada `timepoint` se encuentra en medio de dos `timebound`. Esta segunda coordenada del tiempo nos sirve porque es en esta coordenada donde se calculan los cambios  y por lo general en este punto del tiempo tendremos medidas de tasa de cambio de cantidades. 

Lo más importante es ver que la variable de emisiones, `emissions`, está definida sobre un `timepoint` y todo lo demás sobre `timebound`.

El número de `timebound` es uno más que el número de `timepoint`s, ya que el intervalo incluye el punto final y el punto inicial. 



Para definir el tiempo se utiliza el atributo de `define_time` que toma argumentos muy similares a `np.arange`

```python
# crear horizonte de tiempo de 1950 a 2050 en saltos 1 año
f.define_time(1950, 2050, 1)
# timebounds
print(f.timebounds)
# timepoints
print(f.timepoints)
```

## **<font color="ForestGreen"> Paso 5: Definir las especies químicas </font>**

FaIR es un modelo cuya principal ecuación es el ciclo de los gases. Los gases se emiten, interactúan químicamente y su concentración aumenta o decade dependiendo de este fino balance. Las ecuaciones para el ciclo del carbono, y el ciclo de los demás gases está definido para cada cajita atmosférica.

\begin{align}
\frac{dR_i(t)}{dt}=a_i E(t) - \frac{R_i(t)}{\alpha(t)\tau_i}
\end{align}

donde $R$ es la cantidad del agente o especie en la caja atmosférica $i$, $a_i$ es la fracción de emisiones entrando a esta caja $i$, $\tau_i$ es el tiempo de vida media en años para ese gas en esa caja, $E(t)$ es la serie de tiempo de emisiones emitidas a la atmósfera y $\alpha(t)$ es un coeficiente de ajuste de la concentración en cada caja. 

La concentración total ($C(t)$) de cada agente en toda la atmósfera es entonces la suma de la cantidad $R$ en cada caja: 

\begin{align}
C(t)=C_0 + \sum_{i=1}^n R_i(t)
\end{align}


Esto define los forzantes, antropogénicos o naturales que van a ser incluidos en cada escenario. 
Una `specie` es algo que puede ser directamente emitido como el CO$_2$ a través de combustibles fósiles, o puede ser una categoría donde el forzamiento tiene que ser calculado de emisiones precursoras como las interacciones nube-aerosol. 

Cada `specie` está asignada a un nombre que se utiliza para distinguirlo de otras especies. 
Además vamos a asignarle propiedades a cada especie. La lista completa de especies químicas disponibles es:  “co2 ffi”, “co2 afolu”, “co2”, “ch4”, “n2o”, “cfc-11”, “other halogen”, “f-gas”, “sulfur”, “black carbon”, “organic carbon”, “other slcf”, “nox aviation”, “eesc”, “ozone”, “ari”, “aci”, “contrails”, “lapsi”,”h2o stratospheric”, “land use”, “volcanic”, “solar”, “unspecified”.

#### **<font color="ForestGreen">Paso 7d: configuraciones climáticas  </font>**

Ahora el paso más truculento, rellenar los valores climáticos necesarios, como lo es el calor específico del océano en cada capa, las constantes de transferencia de calor, etc. 
Para esto necesitamos cargar primero para cada modelo sus parámetros, que alguien más afortunadamente ya calculó para nosotros y que nosotros guardamos en nuestro objeto de Pandas `df`. 

Noten bien que el parámetro "seed" se refiere al parámetro de condiciones iniciales y nos dice algo sobre cuántas condiciones iniciales diferentes se intentan para cada modelo y cada simulación.

```python
# Parámetro estocastico
seed = 1355763

# Para cada configuracion tenemos que establecer los parametros, entonces lo hacemos en un loop
for config in configs:
    model, run = config.split('_')
    condition = (df['model']==model) & (df['run']==run)
    fill(f.climate_configs['ocean_heat_capacity'], df.loc[condition, 'C1':'C3'].values.squeeze(), config=config)
    fill(f.climate_configs['ocean_heat_transfer'], df.loc[condition, 'kappa1':'kappa3'].values.squeeze(), config=config)
    fill(f.climate_configs['deep_ocean_efficacy'], df.loc[condition, 'epsilon'].values[0], config=config)
    fill(f.climate_configs['gamma_autocorrelation'], df.loc[condition, 'gamma'].values[0], config=config)
    fill(f.climate_configs['sigma_eta'], df.loc[condition, 'sigma_eta'].values[0], config=config)
    fill(f.climate_configs['sigma_xi'], df.loc[condition, 'sigma_xi'].values[0], config=config)
    fill(f.climate_configs['stochastic_run'], True, config=config)
    fill(f.climate_configs['use_seed'], True, config=config)
    fill(f.climate_configs['seed'], seed, config=config)
    
    seed = seed + 399
```

```python
def lineplot_setup(xmin, xmax, xlab, ylab, ymin=None, ymax=None, title=None, legend=False):
    """
    Configura el entorno gráfico para un diagrama de líneas.

    Este método permite establecer los límites de los ejes, las etiquetas, 
    el título y, opcionalmente, activar la leyenda para un gráfico de líneas.

    Parámetros:
    -----------
    xmin : float
        Valor mínimo del eje X.
    xmax : float
        Valor máximo del eje X.
    xlab : str
        Etiqueta para el eje X.
    ylab : str
        Etiqueta para el eje Y.
    ymin : float, opcional
        Valor mínimo del eje Y (por defecto es None, no se ajusta).
    ymax : float, opcional
        Valor máximo del eje Y (por defecto es None, no se ajusta).
    title : str, opcional
        Título del gráfico (por defecto es None, no se asigna título).
    legend : bool, opcional
        Activa la leyenda del gráfico si se establece en True (por defecto es False).

    Retorna:
    --------
    None
        La función no retorna nada, sólo ajusta el gráfico.

    Ejemplo de uso:
    ---------------
    lineplot_setup(0, 10, 'Tiempo (s)', 'Velocidad (m/s)', 0, 100, title='Gráfico de Velocidad', legend=True)
    """
    
    # Especificar límites mínimos y máximos del gráfico
    plt.xlim([xmin, xmax])
    plt.ylim([ymin, ymax])

    # Cuadrícula interna del gráfico
    plt.grid(linestyle='--', alpha=0.66)
    plt.minorticks_on()

    # Ajustar el tamaño de las etiquetas de ambos ejes
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.xlabel(xlab, fontsize=14)
    plt.ylabel(ylab, fontsize=14)
    plt.title(title, fontsize=15)

    # Mostrar la leyenda si está activada
    if legend:
        plt.legend(fancybox=True, shadow=True, fontsize=12.5)
```

<h1>Función <code>lineplot_setup</code></h1>

<p>La función <code>lineplot_setup</code> es una herramienta que permite configurar un gráfico de líneas en <code>matplotlib</code>. A continuación se explica cada uno de sus parámetros y funcionalidades.</p>

<h2>Parámetros</h2>

<ul>
    <li><strong>xmin</strong> (float): Valor mínimo del eje X.</li>
    <li><strong>xmax</strong> (float): Valor máximo del eje X.</li>
    <li><strong>xlab</strong> (str): Etiqueta para el eje X.</li>
    <li><strong>ylab</strong> (str): Etiqueta para el eje Y.</li>
    <li><strong>ymin</strong> (float, opcional): Valor mínimo del eje Y. Si no se especifica, no se ajusta.</li>
    <li><strong>ymax</strong> (float, opcional): Valor máximo del eje Y. Si no se especifica, no se ajusta.</li>
    <li><strong>title</strong> (str, opcional): Título del gráfico. Si no se especifica, no se mostrará ningún título.</li>
    <li><strong>legend</strong> (bool, opcional): Si es <code>True</code>, se mostrará la leyenda en el gráfico. El valor predeterminado es <code>False</code>.</li>
</ul>

<h2>Funcionalidades</h2>

<p>La función ajusta varios aspectos del gráfico, como los límites de los ejes, la cuadrícula interna y el tamaño de las etiquetas de los ejes. Además, permite añadir un título al gráfico y activar la leyenda si es necesario.</p>
