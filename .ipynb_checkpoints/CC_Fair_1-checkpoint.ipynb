## **<font color="OrangeRed"> Ejemplo de una clase </font>**
En programaci칩n orientada a objetos, una clase es una plantilla o modelo que define las caracter칤sticas y comportamientos de un conjunto de objetos. Un objeto es una instancia espec칤fica de una clase, que tiene los atributos y m칠todos definidos en esa plantilla.

Imagina que una clase es como el dise침o arquitect칩nico de una casa: describe c칩mo deber칤a ser una casa, pero no es una casa en s칤. Cuando se construye una casa basada en ese dise침o, entonces tenemos un objeto que sigue el plano original.

En el contexto de una clase universitaria, podemos pensar en una clase de programaci칩n que modele el comportamiento y la informaci칩n de una asignatura. Esta clase podr칤a incluir atributos como los nombres de los estudiantes, sus equipos de trabajo y sus calificaciones. Adem치s, podr칤a tener m칠todos para agregar estudiantes, asignar calificaciones y mostrar la informaci칩n de los participantes.

El siguiente paso es iniciar la clase, lo que en t칠rminos de programaci칩n significa crear una instancia de esa clase y guardarla en una variable, para poder manipularla y utilizar sus m칠todos.

```python
# Definici칩n de la clase ClaseUniversitaria
class ClaseUniversitaria:
    def __init__(self, nombre_clase):
        self.nombre_clase = nombre_clase
        self.estudiantes = {}  # Diccionario para almacenar nombres de estudiantes y sus equipos
        self.calificaciones = {}  # Diccionario para almacenar nombres de estudiantes y sus calificaciones
        print(f"Clase '{self.nombre_clase}' iniciada.")

    def agregar_estudiante(self, nombre_estudiante, equipo):
        # Agrega un estudiante con su equipo de trabajo
        self.estudiantes[nombre_estudiante] = equipo
        self.calificaciones[nombre_estudiante] = None  # Inicialmente, no hay calificaci칩n
        print(f"Estudiante '{nombre_estudiante}' agregado al equipo '{equipo}'.")

    def asignar_calificacion(self, nombre_estudiante, calificacion):
        # Asigna una calificaci칩n a un estudiante
        if nombre_estudiante in self.calificaciones:
            self.calificaciones[nombre_estudiante] = calificacion
            print(f"Calificaci칩n {calificacion} asignada a '{nombre_estudiante}'.")
        else:
            print(f"Estudiante '{nombre_estudiante}' no encontrado en la clase.")

    def mostrar_informacion(self):
        # Muestra la lista de estudiantes, equipos y calificaciones
        print(f"Informaci칩n de la clase '{self.nombre_clase}':")
        for estudiante, equipo in self.estudiantes.items():
            calificacion = self.calificaciones[estudiante]
            print(f"Estudiante: {estudiante}, Equipo: {equipo}, Calificaci칩n: {calificacion}")

# Paso 1: Iniciar la clase
clase = ClaseUniversitaria("F칤sica del Clima")

# Agregar algunos estudiantes y asignar equipos
clase.agregar_estudiante("Ana P칠rez", "Equipo 1")
clase.agregar_estudiante("Carlos L칩pez", "Equipo 2")

# Asignar calificaciones a los estudiantes
clase.asignar_calificacion("Ana P칠rez", 95)
clase.asignar_calificacion("Carlos L칩pez", 88)

# Mostrar la informaci칩n de la clase
clase.mostrar_informacion()
```

Es importante mencionar en este punto que FaIR es un modelo de cajas, en este caso una "layer" se refiere a las capas, la primera es la superficie, la segunda es la primer capa oc칠anica y la tercera es el oc칠ano profundo. Esto en su definici칩n por default son 3 "layers". Los "gasboxes" son las cajas atmosf칠ricas donde se guardan los gases y donde se resuelve la temperatura. La caja m치s inferior es la superficial.

## **<font color="ForestGreen"> Paso 2 Tiempo de simulacion </font>**

En FaIR el tiempo es la 칰nica dimensi칩n y por lo tanto, t칠cnicamente podr칤amos hacer la simulaci칩n que queramos en el tiempo que queramos. De manera general, vamos a querer hacer simulaciones que sean parecidas a las del IPCC. 

Hay dos indicadores diferentes del tiempo en este modelo: `timebound` y `timepoint`. 
El intervalo de tiempo `timebound` son las coordenadas del tiempo que rodean o que se encuentran alrededor de un paso de tiempo, son fotos instant치neas de lo que est치 sucediendo. El punto en el tiempo, `timepoint` es lo que sucede en el intervalo de tiempo. Es decir, cada `timepoint` se encuentra en medio de dos `timebound`. Esta segunda coordenada del tiempo nos sirve porque es en esta coordenada donde se calculan los cambios  y por lo general en este punto del tiempo tendremos medidas de tasa de cambio de cantidades. 

Lo m치s importante es ver que la variable de emisiones, `emissions`, est치 definida sobre un `timepoint` y todo lo dem치s sobre `timebound`.

El n칰mero de `timebound` es uno m치s que el n칰mero de `timepoint`s, ya que el intervalo incluye el punto final y el punto inicial. 



Para definir el tiempo se utiliza el atributo de `define_time` que toma argumentos muy similares a `np.arange`

```python
# crear horizonte de tiempo de 1950 a 2050 en saltos 1 a침o
f.define_time(1950, 2050, 1)
# timebounds
print(f.timebounds)
# timepoints
print(f.timepoints)
```

## **<font color="ForestGreen"> Paso 5: Definir las especies qu칤micas </font>**

FaIR es un modelo cuya principal ecuaci칩n es el ciclo de los gases. Los gases se emiten, interact칰an qu칤micamente y su concentraci칩n aumenta o decade dependiendo de este fino balance. Las ecuaciones para el ciclo del carbono, y el ciclo de los dem치s gases est치 definido para cada cajita atmosf칠rica.

\begin{align}
\frac{dR_i(t)}{dt}=a_i E(t) - \frac{R_i(t)}{\alpha(t)\tau_i}
\end{align}

donde $R$ es la cantidad del agente o especie en la caja atmosf칠rica $i$, $a_i$ es la fracci칩n de emisiones entrando a esta caja $i$, $\tau_i$ es el tiempo de vida media en a침os para ese gas en esa caja, $E(t)$ es la serie de tiempo de emisiones emitidas a la atm칩sfera y $\alpha(t)$ es un coeficiente de ajuste de la concentraci칩n en cada caja. 

La concentraci칩n total ($C(t)$) de cada agente en toda la atm칩sfera es entonces la suma de la cantidad $R$ en cada caja: 

\begin{align}
C(t)=C_0 + \sum_{i=1}^n R_i(t)
\end{align}


Esto define los forzantes, antropog칠nicos o naturales que van a ser incluidos en cada escenario. 
Una `specie` es algo que puede ser directamente emitido como el CO$_2$ a trav칠s de combustibles f칩siles, o puede ser una categor칤a donde el forzamiento tiene que ser calculado de emisiones precursoras como las interacciones nube-aerosol. 

Cada `specie` est치 asignada a un nombre que se utiliza para distinguirlo de otras especies. 
Adem치s vamos a asignarle propiedades a cada especie. La lista completa de especies qu칤micas disponibles es:  띾o2 ffi, 띾o2 afolu, 띾o2, 띾h4, 랉2o, 띾fc-11, 랊ther halogen, 랁-gas, 랍ulfur, 락lack carbon, 랊rganic carbon, 랊ther slcf, 랉ox aviation, 란esc, 랊zone, 라ri, 라ci, 띾ontrails, 랇apsi,렫2o stratospheric, 랇and use, 랐olcanic, 랍olar, 랏nspecified.

#### **<font color="ForestGreen">Paso 7d: configuraciones clim치ticas  </font>**

Ahora el paso m치s truculento, rellenar los valores clim치ticos necesarios, como lo es el calor espec칤fico del oc칠ano en cada capa, las constantes de transferencia de calor, etc. 
Para esto necesitamos cargar primero para cada modelo sus par치metros, que alguien m치s afortunadamente ya calcul칩 para nosotros y que nosotros guardamos en nuestro objeto de Pandas `df`. 

Noten bien que el par치metro "seed" se refiere al par치metro de condiciones iniciales y nos dice algo sobre cu치ntas condiciones iniciales diferentes se intentan para cada modelo y cada simulaci칩n.

```python
# Par치metro estocastico
seed = 1355763

# Para cada configuracion tenemos que establecer los parametros, entonces lo hacemos en un loop
for config in configs:
    model, run = config.split('_')
    condition = (df['model']==model) & (df['run']==run)
    fill(f.climate_configs['ocean_heat_capacity'], df.loc[condition, 'C1':'C3'].values.squeeze(), config=config)
    fill(f.climate_configs['ocean_heat_transfer'], df.loc[condition, 'kappa1':'kappa3'].values.squeeze(), config=config)
    fill(f.climate_configs['deep_ocean_efficacy'], df.loc[condition, 'epsilon'].values[0], config=config)
    fill(f.climate_configs['gamma_autocorrelation'], df.loc[condition, 'gamma'].values[0], config=config)
    fill(f.climate_configs['sigma_eta'], df.loc[condition, 'sigma_eta'].values[0], config=config)
    fill(f.climate_configs['sigma_xi'], df.loc[condition, 'sigma_xi'].values[0], config=config)
    fill(f.climate_configs['stochastic_run'], True, config=config)
    fill(f.climate_configs['use_seed'], True, config=config)
    fill(f.climate_configs['seed'], seed, config=config)
    
    seed = seed + 399
```

```python
def lineplot_setup(xmin, xmax, xlab, ylab, ymin=None, ymax=None, title=None, legend=False):
    """
    Configura el entorno gr치fico para un diagrama de l칤neas.

    Este m칠todo permite establecer los l칤mites de los ejes, las etiquetas, 
    el t칤tulo y, opcionalmente, activar la leyenda para un gr치fico de l칤neas.

    Par치metros:
    -----------
    xmin : float
        Valor m칤nimo del eje X.
    xmax : float
        Valor m치ximo del eje X.
    xlab : str
        Etiqueta para el eje X.
    ylab : str
        Etiqueta para el eje Y.
    ymin : float, opcional
        Valor m칤nimo del eje Y (por defecto es None, no se ajusta).
    ymax : float, opcional
        Valor m치ximo del eje Y (por defecto es None, no se ajusta).
    title : str, opcional
        T칤tulo del gr치fico (por defecto es None, no se asigna t칤tulo).
    legend : bool, opcional
        Activa la leyenda del gr치fico si se establece en True (por defecto es False).

    Retorna:
    --------
    None
        La funci칩n no retorna nada, s칩lo ajusta el gr치fico.

    Ejemplo de uso:
    ---------------
    lineplot_setup(0, 10, 'Tiempo (s)', 'Velocidad (m/s)', 0, 100, title='Gr치fico de Velocidad', legend=True)
    """
    
    # Especificar l칤mites m칤nimos y m치ximos del gr치fico
    plt.xlim([xmin, xmax])
    plt.ylim([ymin, ymax])

    # Cuadr칤cula interna del gr치fico
    plt.grid(linestyle='--', alpha=0.66)
    plt.minorticks_on()

    # Ajustar el tama침o de las etiquetas de ambos ejes
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.xlabel(xlab, fontsize=14)
    plt.ylabel(ylab, fontsize=14)
    plt.title(title, fontsize=15)

    # Mostrar la leyenda si est치 activada
    if legend:
        plt.legend(fancybox=True, shadow=True, fontsize=12.5)
```
